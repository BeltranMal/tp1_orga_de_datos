```python
preferencias = costumer_data.iloc[:,8:12]
preferencias = preferencias.rename(columns={"booking_origin": "country"})
```

```python

def corregir_rates(serie):
  if(serie == "None"):
    return -1
  elif(serie == "\n\t\t\t\t\t\t\t\t\t\t\t\t\t5"):
    return 5
  else:
    return serie

airline_data["rates"] = airline_data["rates"].map(corregir_rates)
airline_data.rates = airline_data.rates.astype(np.int8)


```

```python
datos_filtrados = airline_data.iloc[:,[2,4]]
datos_filtrados
```

```python

preferencias_paises = pd.merge(datos_filtrados,preferencias,how = "outer")
promedios_altos =preferencias_paises[ preferencias_paises.rates > 8 ]#consideramos que 9 es un puntaje alto
promedios_altos_filtrados = promedios_altos.dropna()
promedios_altos_filtrados
```
